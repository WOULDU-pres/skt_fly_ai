### groupby()

```python
df.groupby('sex')['survived'].value_counts()
```

```python
df.groupby(['sex','pclass'])['survived'].value_counts()
```

- apply 함수

```python
def myfunc(x):
    if x == 'male':
        return 1
    else:
        return 0
```

```python
df['sex_encoded'] = df['sex'].apply(myfunc)
df
```

```python
def survived(x):
    if x == "saved":
        return 1
    else:
        return 0
```

```python
df['survived'] = df['survived'].apply(survived)
df
```

```python
df.groupby(['sex','pclass'])['survived'].sum()
```
