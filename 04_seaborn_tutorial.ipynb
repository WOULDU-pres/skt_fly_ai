```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.__version__
```

```python
df = sns.load_dataset('penguins')
df
```

```python
df.info()
```

```python
df['species'].value_counts()
```

```python
df.groupby(['island','species','sex'])['flipper_length_mm'].value_counts()
```

```python
df.columns
```

### Axis Level

```python
plt.figure(figsize=(4,3))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species')
plt.show()
```

```python
plt.figure(figsize=(7,3))
sns.histplot(data=df, x='species', hue='species')
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,5))
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax=axes[0,0])
sns.histplot(data=df, x='species', hue='species', ax=axes[1,0])
sns.scatterplot(data=df, x='flipper_length_mm', y='bill_length_mm', hue='species', ax=axes[0,1])
sns.histplot(data=df, x='species', hue='species', ax=axes[1,1])
```

### figure level

```python
df.columns
```

```python
sns.displot(data=df, 
            x='flipper_length_mm', 
            kind='hist', 
            hue='species',
            col='species')
```

```python
sns.displot(data=df, 
            x='flipper_length_mm', 
            kind='kde', 
            hue='species',
            col='species')
```

```python
sns.displot(data=df, 
            x='flipper_length_mm', 
            kind='ecdf', 
            hue='species',
            col='species')
```

```python
df.columns
```

```python
sns.relplot(data=df, 
            x='flipper_length_mm',
            y='bill_length_mm', 
            hue ='species',
            col='species',
            height=5,
            aspect=1)
```
