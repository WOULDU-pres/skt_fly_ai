```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tips.csv')
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df['time'].value_counts()
```

```python
df.columns
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='size')
```

```python
sns.histplot(data=df,  x='tip', hue='sex')
```

```python
fig, ax = plt.subplots(2,2)
ax = ax.reshape(-1)
sns.scatterplot(data=df, x='total_bill', y='tip', hue='size', ax=ax[0])
sns.histplot(data=df, x='tip', hue ='time', ax=ax[1])
```

```python
sns.relplot(data=df,
           x='total_bill',
           y='time',
           hue = 'sex',
           col='sex'
           )
```
