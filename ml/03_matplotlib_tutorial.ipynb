### 산정도 

```python
# !gdown https://raw.githubusercontent.com/devdio/datasets/main/iris.csv
```

```python
df = pd.read_csv('iris.csv', index_col = 0)
df.shape
```

```python
df.head()
```

```python
plt.scatter(x = df['SepalLengthCm'],
           y = df['SepalWidthCm'])
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title('iris')
```

```python
df['Species'].unique()
```

```python
plt.scatter(x = df.loc[df["Species"] == 'Iris-setosa']['SepalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-setosa']['SepalWidthCm'])
plt.scatter(x = df.loc[df["Species"] == 'Iris-versicolor']['SepalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-versicolor']['SepalWidthCm'], color ='g')
plt.scatter(x = df.loc[df["Species"] == 'Iris-virginica']['SepalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-virginica']['SepalWidthCm'],color ='r')
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title('iris')
```

```python
plt.scatter(x = df.loc[df["Species"] == 'Iris-setosa']['PetalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-setosa']['PetalWidthCm'])
plt.scatter(x = df.loc[df["Species"] == 'Iris-versicolor']['PetalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-versicolor']['PetalWidthCm'], color ='g')
plt.scatter(x = df.loc[df["Species"] == 'Iris-virginica']['PetalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-virginica']['PetalWidthCm'],color ='r')
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.title('iris')
```

```python
x1 = df.loc[df["Species"] == 'Iris-setosa']['SepalLengthCm']
y1 = df.loc[df["Species"] == 'Iris-setosa']['SepalWidthCm']
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)


ax1.scatter(x = x1, y = y1)
ax1.scatter(x = df.loc[df["Species"] == 'Iris-versicolor']['SepalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-versicolor']['SepalWidthCm'], color ='g')
ax1.scatter(x = df.loc[df["Species"] == 'Iris-virginica']['SepalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-virginica']['SepalWidthCm'],color ='r')

ax2.scatter(x = df.loc[df["Species"] == 'Iris-setosa']['PetalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-setosa']['PetalWidthCm'])
ax2.scatter(x = df.loc[df["Species"] == 'Iris-versicolor']['PetalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-versicolor']['PetalWidthCm'], color ='g')
ax2.scatter(x = df.loc[df["Species"] == 'Iris-virginica']['PetalLengthCm'],
           y = df.loc[df["Species"] == 'Iris-virginica']['PetalWidthCm'],color ='r')


ax1.set_title("Iris Sepal")
ax2.set_title("Iris Petal")
plt.show()
```

```python
mydata = [[1,2,3,4],[10,11,12,13], [20,25,43,54],[32,35,34,32]]
fig, ax = plt.subplots(2,2)
   
ax = ax.reshape(-1)
for i in range(4):
    ax[i].plot(mydata[i])
```

```python
fig, ax = plt.subplots(2,2)
ax = ax.reshape(-1)
for x, z in zip(ax, mydata):
    x.plot(z)
```
